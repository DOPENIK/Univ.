```python
import numpy as np
from imageio import imread
import matplotlib.pyplot as plt

img = plt.imread('dog.jpg')
img_tinted = img*[1,0.8,0.8]

plt.subplot(1,2,1)
plt.imshow(img)

plt.subplot(1,2,2)

plt.imshow(np.uint8(img_tinted))
plt.show()
```
