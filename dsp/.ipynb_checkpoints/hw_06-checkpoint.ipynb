```python
# Get thinkdsp.py

import os

if not os.path.exists('thinkdsp.py'):
    !wget https://github.com/AllenDowney/ThinkDSP/raw/master/code/thinkdsp.py
        
import thinkdsp
from thinkdsp import decorate
import numpy as np
```

```python
np.random.seed(17)
```

```python
wave = thinkdsp.read_wave('forest_mumur.wav')
segment = wave.segment(start=0,duration=2)
segment.make_audio()
```

frequency 확인

```python
spectrum = wave.make_spectrum()
spectrum.plot(linewidth=0.5)
decorate(xlabel='Frequency (Hz)',
         ylabel='Amplitude')
```

power spectrum 만들기 

```python
spectrum.plot_power(linewidth=0.5)
decorate(xlabel='Frequency (Hz)',
         ylabel='Power')
```

Brownian의 형태인걸로 확인 됨 , Log log scale로 확인

```python
spectrum.hs[0] = 0

spectrum.plot_power(linewidth=0.5)

loglog = dict(xscale='log', yscale='log')
decorate(xlabel='Frequency (Hz)',
         ylabel='Power',
         **loglog)
```

```python
from thinkdsp import BrownianNoise

signal =  BrownianNoise()
wave = signal.make_wave(duration=0.5, framerate=11025)
spectrum = wave.make_spectrum()
result = spectrum.estimate_slope()
result.slope
```

slope가 -2에 수렴하고 시간에 따른 Frequency 모양이 음의 방향으로 계속 가는 것을 볼 수 있다. 결과적으로 Brownian 소음처럼 보인다.
