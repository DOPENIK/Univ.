```python
import thinkdsp

wave = thinkdsp.read_wave('181425__serylis__guitar-chord.wav')
segment = wave.segment(start=0,duration=0.5)
segment.make_audio()
```

```python
spectrum = segment.make_spectrum()
spectrum.plot_power()
```

```python
spectrum.plot_power()
thinkplot.config(xscale='log',yscale='log')
```

```python
brownWave = thinkdsp.BrownianNoise().make_wave(duration = 1.0, framerate = 11025)
brownSpectrum = brownWave.make_spectrum()

spectrum.plot_power()
brownSpectrum.plot_power(color = 'red')

thinkplot.config(xscale='log',yscale='log')
```
