```python
wave = thinkdsp.read_wave('181425__serylis__guitar-chord.wav')
segment = wave.segment(start=0,duration=0.5)
segment.make_audio()
```

```python
print('Number of samples', len(wave.ys))
print('Timestep in ms', 1 / wave.framerate * 1000)
```

```python
segment.plot()
decorate(xlabel='Time (s)')
```

```python
wave.normalize()
wave.apodize() # denoise    .
wave.plot()
decorate(xlabel='Time (s)')
```
